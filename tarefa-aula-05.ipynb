```python
import pandas as pd
import numpy as np
```

Tratar as colunas sem informação adicionando o texto “Não informado”.

```python
df = pd.read_csv('/content/dataset_topyoutube.csv')

df = df.fillna('Não informado')
df.isna().sum()
```

Formatar a coluna de total de inscritos multiplicando por 100 milhões.

```python
df['Total Views'] = df['Total Views'].apply(lambda x: x.replace(',', ''))
```

```python
df['Total Views'] = df['Total Views'].astype(float)
```

```python
df['Total Views'] = df['Total Views'] * 1e8
df
```

Formatar coluna “AVG” para 2 casas decimais depois da vírgula.

```python
df['Avg'] = round(df['Avg'], 2)
df
```

Mostrar os top 10 usuários.

```python
df.sort_values(by='Total Views', ascending=False).head(10)

```

Mostrar primeiros 100 usuários por nome decrescente

```python
df.sort_values(by='Artist', ascending=False).head(100)
```

Salvar um novo CSV com a informação tratada.

```python
df.to_csv('/content/dataset2_topyoutube.csv')
```
